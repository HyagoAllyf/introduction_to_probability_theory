```python
import matplotlib.pyplot as plt
import numpy as np
rate = 5 # taxa
n = 30 # numero de eventos
T = np.zeros(n) # vetor que armazena os tempos entre eventos
te = 0
for i in range(n):
    ti = np.random.exponential(1/rate)
    te = te + ti
    T[i] = te
plt.figure(figsize=(8,2))
plt.title("Tempo de cada evento", fontdict={'fontname': 'Times New Roman', 'fontsize':21}, y = 1.03)
for t in T:
    plt.axvline(x=t)
plt.show(True)
```

Podemos ainda obter do n√∫mero de eventos em cada intervalo de tempo com comprimento *t*. A probabilidade de ocorrer *k* eventos em um
$$ P[N(t)=k] = \dfrac{e^{-\lambda t}(\lambda t)^t}{k!}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math as math

rate = 5
tmax = 20
nsim = 10000
Ns = []
for s in range(0, nsim):
    te = 0
    ne = 0
    while (te < tmax):
        te = te + np.random.exponential(1/rate)
        ne = ne + 1
    Ns.append(ne)
plt.figure(figsize=(8,6))
plt.hist(x=Ns, bins = 'auto', color= '#0504aa', alpha=0.7, rwidth=0.85, density = True, label='Simula√ß√£o')
plt.xlabel(r'$N(t)$', fontsize=20)
plt.ylabel(r'$P[N(t)]$', fontsize=20)

x = range(np.min(Ns), np.max(Ns))
P = []
for k in x:
    prob = (np.exp(-rate*tmax)*(rate*tmax)**k)/math.factorial(k)
    P.append(prob)
plt.plot(x,P, color='red', label='Curva Teorica')
plt.legend()
plt.show(True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = ['red', 'blue', 'green']
rates = [2, 5, 10]
plt.figure(figsize=(10,6))
aux = 0
for rate in rates:
    N = []
    T = []
    ne = 0
    te = 0
    while (te < tmax):
        te = te + np.random.exponential(1/rate)
        ne = ne + 1
        N.append(ne)
        T.append(te)
    plt.step(T, N, color = colors[aux], label = "Lambda = %d"%rate)
    aux = aux + 1
plt.legend()
plt.xlabel(r'$t$', fontsize=20)
plt.ylabel(r'$N(t)$', fontsize=20)
plt.show(True)
```
